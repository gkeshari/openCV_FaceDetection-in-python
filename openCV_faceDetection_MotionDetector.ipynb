# Face Detection

```python
# image --> create a cascade classifier, it will contain features of the face(it will determine where is the face)
# --> opencv --> numpy array(search for row n column values of the face (face rectangle coordinates))
# --> display the image with rectangular face box
```

```python
import cv2

#cv2.CascadeClassifier.load("haarcascade_frontalface_default.xml")

# create a CascadeClassifier object
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

face_cascade.load("haarcascade_frontalface_default.xml")
```

```python
# Reading the image as it is
image = cv2.imread("C:\\Users\\Brij\\Desktop\\govind madhav keshari.jpg",1)

#Reading the image as gray scale image 
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Search for the face rectangle coordinates of the image
faces = face_cascade.detectMultiScale(gray_img,scaleFactor=1.05,minNeighbors=5)

print(type(faces))
print(faces)  # coordinates of the face


# same thing to detect eyes
eyes = eye_cascade.detectMultiScale(gray_img)

for ex,ey,ew,eh in eyes:
    cv2.rectangle(image, (ex,ey), (ex+ew,ey+eh), (255,0,0), 2)
    
    

# for adding rectangle to face we r using for loop
for (x,y,w,h) in faces:
    image = cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 3)
    
cv2.imshow('Gray', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# lets add a window that shows the video   (for above cell)

# check : is a bool data type, returns true if Python is able to read the VideoCapture object

# frame : is a Numpy array, it represents the first image that video captures
```

```python
# till now we just captured the first frame 
# how to capture the video, instead of First image/Frame of the video??
# for that we r going to use a loop

# in order to capture the video, we will be using the 'while' loop. While condition will be such that,
# ultil unless 'check' is True, Python will display the frames.
```

```python
import cv2,time

video= cv2.VideoCapture(0)
a=1

while True:
    a=a+1
    check,frame=video.read()
    print(frame)
    gray= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # convert the frame into a gray scale image
    cv2.imshow('capturing',gray)
    key=cv2.waitKey(1)  # this will generate a new frame after every 1 milisec.
    if key==ord('q'):   # once we enter q the window will be destroyed
        break
        
print(a)  # this will print the no. of frames
video.release()
cv2.destroyAllWindows()
```

```python
# Use case : Motion Detector
```

```python
# problem statement : u hav been approached a company that is studying human behaviour. ur task is to give them a webcam
#  that can detect the motion or any movement in front of it. This should return a graph, this graph should contain for 
# how long the human/object was in front of the camera.
```

```python
# flow diagram of use case:
# start--->save the initial image in a frame--->convert this image to a gaussian blur image--->
# take the frames with the object & convert it into gaussian blur image--->cal. the diffc--->
# define the threshold to remove the shadows & other noises--->define the borders of the object--->
# add the rectangular box around the object--->cal. the time, when object appears and exits the frame 
```

```python
import cv2,time,pandas

first_frame = None
video = cv2.VideoCapture(0)
while True:
    check,frame = video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray=cv2.GaussianBlur(gray,(21,21),0)
    
    if first_frame is None:
        first_frame=gray
        continue

    delta_frame = cv2.absdiff(first_frame,gray)
    # cal. the diffrnc between first frame and other frames 


    thresh_delta = cv2.threshold(delta_frame, 30, 255, cv2.THRESH_BINARY)[1]
    thresh_delta = cv2.dilate(thresh_delta,None,iterations=0)
    # provides a threshold value, such that it will convert the diffrence value less than 30 to black. 
    # if the diffrnce is greater than 30 it will convert those pixels to white



    cnts,_ = cv2.findContours(thresh_delta.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    # define the contour area basically, add the borders



    for contour in cnts:
        if cv2.contourArea(contour)<1000:
            continue
        x,y,w,h=cv2.boundingRect(contour)
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0),3)
    

    
    cv2.imshow('Frame',frame)
    cv2.imshow('Capturing',gray)
    cv2.imshow('delta',delta_frame)
    cv2.imshow('thresh',thresh_delta)

    key = cv2.waitKey(1)
    if key==ord('q'):
        break
video.release()
cv2.destroyAllWindows()


        
```
