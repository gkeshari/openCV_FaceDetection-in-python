# L7 : color filtering

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# filtering for specific color
```

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    # hsv : hue sat value
    lower_red = np.array([150,15,0])
    upper_red = np.array([180,255,255])
    
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    
    key = cv2.waitKey(3)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```

# L8 : Blurring & smoothening

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    # hsv : hue sat value
    lower_red = np.array([150,15,0])
    upper_red = np.array([180,255,255])
    
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    kernel = np.ones((15,15), np.float32)/225
    smooth = cv2.filter2D(res, -1, kernel)
    
    gaussianBlur =  cv2.GaussianBlur(res, (15,15), 0)
    
    medianBlur = cv2.medianBlur(res,15)
    
    bilateral = cv2.bilateralFilter(res,15,75,75) # useless
    
    
    cv2.imshow('frame',frame)
    #cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    cv2.imshow('smooth',smooth)
    cv2.imshow('Gaussian',gaussianBlur)
    cv2.imshow('medianBlur',medianBlur)
    cv2.imshow('bilateral',bilateral)
    
    key = cv2.waitKey(3)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```

# L9 : Morphological Transformations

```python
# to remove noise

# first 2 morphological tranformations are : 'erosion' , 'dilation'

# two more versions : 'opening' & 'closing'
# opening : to remove false positives (remove from background)
# closing : '' ''      ''   negatives

# two more versions are : 'Tophat' & 'Blackhat'
```

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    # hsv : hue sat value
    lower_red = np.array([150,15,0])
    upper_red = np.array([180,255,255])
    
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    kernel = np.ones((5,5), np.uint8)
    erosion = cv2.erode(mask, kernel, iterations = 1)
    dilation = cv2.dilate(mask, kernel, iterations = 1)
    
    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    
    
    cv2.imshow('frame',frame)
    cv2.imshow('res',res)
    cv2.imshow('erosion',erosion)
    cv2.imshow('dilation',dilation)
    cv2.imshow('opening',opening)
    cv2.imshow('closing',closing)
   
    
    key = cv2.waitKey(3)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```

# L10 : gradient & edge detection

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    
    laplacian = cv2.Laplacian(frame, cv2.CV_64F)  # laplacian gradient
    
    Sobelx = cv2.Sobel(frame, cv2.CV_64F, 1, 0, ksize=5) # both gives directional intensity
    Sobely = cv2.Sobel(frame, cv2.CV_64F, 0, 1, ksize=5)  # y intensity gradient
    
    # edge detactor : Canny edge detactor
    edges = cv2.Canny(frame , 100, 100)


    
    
    cv2.imshow('frame',frame)
    cv2.imshow('laplacian',laplacian)
    cv2.imshow('Sobelx',Sobelx)
    cv2.imshow('Sobely',Sobely)
    cv2.imshow('edges',edges)

   
    
    key = cv2.waitKey(3)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```

# L11 : Template matching

```python
import cv2 
import numpy as np 

img_bgr = cv2.imread('C:\\Users\\Brij\\Desktop\\img086.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

template = cv2.imread('C:\\Users\\Brij\\Desktop\\Govind_Signature.jpeg',0)
w,h = template.shape[::-1]

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.80
loc = np.where(res>= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img_bgr, pt, (pt[0]+w, pt[1]+h), (0,255,255), 2)
    
cv2.imshow('detected', img_bgr)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# L12 : GrabCut Foreground extraction

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('C:\\Users\\Brij\\Desktop\\govind madhav keshari.jpg')
mask = np.zeros(img.shape[:2], np.uint8)

bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)

rect = (40,50,400,500)

cv2.grabCut(img,  mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2) | (mask==0), 0, 1).astype('uint8')
img = img*mask2[:,:,np.newaxis]

plt.imshow(img)
plt.colorbar()
plt.show()
```

# L14 : Feature matching (homography)

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img1 = cv2.imread('C:\\Users\\Brij\\Desktop\\img086.jpg',0)
img2 = cv2.imread('C:\\Users\\Brij\\Desktop\\Govind_Signature.jpeg',0)

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], None, flags=2)
plt.imshow(img3)
plt.show()


```

# L16 : HaarCascade Object detection Face & Eye

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

face_cascade = cv2.CascadeClassifier('haarcascade_frontalFace_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)   # default is also working good.
    
    for x,y,w,h in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 3)
        
    eyes = eye_cascade.detectMultiScale(gray, 1.3, 5)       # default is also working good.
    for x,y,w,h in eyes:
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
        
    cv2.imshow('img', img)
    key = cv2.waitKey(5)
    if key == ord('q'):
        break
cap.release() 
cv2.destroyAllWindows()
```
